```python
def apply_age_time(num):
    if 1930 <= num <= 1940:
        return 1
    elif 1940 < num <= 1950:
        return 2
    elif 1950 < num <= 1960:
        return 3
    elif 1960 < num <= 1970:
        return 4
    elif 1970 < num <= 1980:
        return 5
    elif 1980 < num <= 1990:
        return 6
    elif 1990 < num <= 2000:
        return 7
    elif 2000 < num <= 2010:
        return 8
    elif 2010 < num <= 2019:
        return 9

person_score["year_group"] = person_score["year_of_birth"].apply(apply_age_time)
```

```python
person_score.groupby("year_group")["status"].agg(["count", "mean"])
```
