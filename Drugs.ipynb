```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
```

```python
drug = pd.read_csv("C:\\Users\\ASUS\\Desktop\\covid_pure\\training\\drug_exposure.csv")
gs = pd.read_csv("C:\\Users\\ASUS\\Desktop\\covid_pure\\training\\goldstandard.csv")
data_dict = pd.read_csv("data_dictionary.csv")
drug.head()
```

```python
columns = ["person_id", "drug_concept_id", "drug_exposure_start_date",
           "drug_exposure_end_date", "refills", "quantity", 
           "route_source_value", "dose_unit_source_value"]
drug = drug[columns]
drug.head()
```

```python
d= {}
values = drug.drug_concept_id.unique()
print(values)
for id_x in values:
    try:
        value = data_dict[data_dict['concept_id'] == id_x]['concept_name'].values[0]
        d[id_x] = value
    except:
        pass
    

drug['drug_concept_id'] = drug['drug_concept_id'].map(d)
drug.head()
    
```

```python
drug['drug_exposure_start_date'] = pd.to_datetime(drug['drug_exposure_start_date'])
drug['drug_exposure_end_date'] = pd.to_datetime(drug['drug_exposure_end_date'])
```

```python
drug['time_of_drug'] = drug['drug_exposure_end_date'] - drug['drug_exposure_start_date']
drug.dtypes
```

```python
drug['year'] = pd.DatetimeIndex(drug['drug_exposure_start_date']).year
drug = drug[drug['year'] >= 2020]
```
